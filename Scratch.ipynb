```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

"""
dissemble pictures of rectangle.
input: row, column
return: count : count of dissembled pictures 
out: saved images named 'temp/boxrow%scol%sup and temp/boxrow%scol%sdown' 
"""
img = cv2.imread("temp/raspi_pic.jpg")
row= 2
col= 3
rows,cols,ch = img.shape
print(img.shape)
pts1 = np.float32([[800,0],[1330,0],[800,340],[1312,340]])
pts2 = np.float32([[0,0],[col*56,0],[0,row*56],[col*56,row*56]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(col*56+1,row*56+1))
# 56* row, 56* col
#height up/down = 7/9

# print(dst.shape)
#set manual

#set automatically
for i in range(0,row):
    for j in range(0,col):
        shoeBox = dst[i*28:(i+1)*28 , j*56:(j+1)*56]
        # 0 is up
        cv2.imwrite('temp/box%s.png'%(i*col+j),shoeBox)
        
# #size of up, down
# sizeOfUpBox=int(56*7/16)
# sizeOfDownBox=int(56*9/16)

# #set automatically
# for i in range(0,row):
#     for j in range(0,col):
#         shoeBox = dst[i*56:(i+1)*56 , j*56:(j+1)*56]
#         # 0 is up
#         cv2.imwrite('temp/boxrow%scol%sup.png'%(i,j),shoeBox[0:sizeOfUpBox, 0:56])
#         # 1 is down
#         cv2.imwrite('temp/boxrow%scol%sdown.png'%(i,j),shoeBox[0:sizeOfDownBox, 0:56])
# #       cv2.imwrite('temp/box%s%sdown.png'%(2*(i*col+j)+k),dst[j*56:(j+1)*56 , i*56:(i+1)*56])


plt.subplot(221),plt.imshow(img),plt.title('Input')
plt.subplot(222),plt.imshow(dst),plt.title('Output')
j=0
i=2
shoeBox = dst[i*28:(i+1)*28 , j*56:(j+1)*56]
plt.subplot(223),plt.imshow(shoeBox),plt.title('O1utput')

plt.show()
```
